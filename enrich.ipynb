```python
import pandas as pd
```

```python
#hep
geo='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE129nnn/GSE129374/matrix/GSE129374_series_matrix.txt.gz'
meta=pd.read_table(geo,nrows=8,skiprows=33,index_col=0).T['!Sample_characteristics_ch1'].str.split('disease state: Cirrhosis_').str[1]
```

```python
df=pd.read_table(geo,nrows=1000000,skiprows=33+36,index_col=0)
```

```python
df['n']=df[meta[meta=='alone'].index].mean(1)
df['c']=df[meta[meta=='with_HCC'].index].mean(1)
```

```python
df['d']=df['c']-df['n']
```

```python
df.sort_values('d')
```

```python
pd.concat([df.loc['cg06488957'],meta],axis=1).groupby('!Sample_characteristics_ch1').mean()
```

```python
crc='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE86nnn/GSE86078/suppl/GSE86078_Sample142-149_normalized.txt.gz'
crc=pd.read_table(crc,index_col=0,usecols=[0]+list(range(1,16,2)))
```

```python
ctr='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE32nnn/GSE32146/matrix/GSE32146_series_matrix.txt.gz'
ctr=pd.read_table(ctr,skiprows=31+35,index_col=0)
```

```python
dg=pd.concat([crc.mean(1).rename('c'),ctr.mean(1).rename('n')],axis=1)
```

```python
dg['d']=dg['c']-dg['n']
```

```python
dg.dropna().sort_values('d').head(5).index
```
